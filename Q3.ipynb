```python
# Creates a dictionary having the entity names as key and their counts as values
# Takes 1 arguement ,i.e. results which is a memory-efficient generator of Results objects when using the streaming mode
def count_entities(results):
        d = {}
        names = model.names # Contains a dictionary of the labels and their entity name  
        for i,j in names.items():
            names[i] = j.capitalize() # Capitalizing the entity name for the sake of output format
            
        for r in results: # Iterating through the results generator where each iterable is a result object
            for c in r.boxes.cls: # Getting the labels/entity names in the image and iterating through it
                key = names[int(c)]
                d[key] = (d.get(key, 0)+1)
        return d
```

```python
import glob
import shutil
import os
from ultralytics import YOLO

# Using Yolov8x.pt model since it has most parameters to modify
model = YOLO("yolov8x.pt")
directory = 'All_Images'# Directory in which all the images are in


data = {}
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
  # checking if it is a real file
    if os.path.isfile(f):
        # Predicting / Getting the entities in the image(filename)
        results = model.predict(source = f,save = True,conf = 0.5, stream=True)
        d = count_entities(results)
        data[filename] = d # Creating a dictionary where key is the image name and value is all the data of the entities within it
        
```

```python

all = []
for i,j in data.items():
    all.extend(list(j.keys())) # List of all the (names)entities together of all the images

all = list(set(all)) # Removing repitition

main = []
for i,j in names.items():
    for val in all:
        if j == val:
            main.append(i)
# Creating main ,i.e. a list of all the entities(labels) together of all the images
```

```python
try:
    os.mkdir("Q3_groups")# Creating the directory for the Q3 folders groups
    for i in main:
        os.mkdir(f"Q3_groups/grp{i+1} - {names[i]}") # Creating folders in the format given in the output
except Exception as e:
    print(f"Folder already is present : Error - {e}")

    
for img,itms in data.items():
    for num,name in names.items():
        if name in itms.keys():
            src = "All_Images" # Source Directory of all the input images
            dst = f"Q3_groups/grp{num+1} - {names[num]}" # Destination Directory 
            for imgfile in glob.iglob(os.path.join(src,f"{img}")):
                shutil.copy(imgfile,dst) # Copying the image from the source directory and pasting it to the destination directory
```
