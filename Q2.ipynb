```python
#create csv function creates the csv file for image and its entities ,given the dictionary of images
# By default path is "Q2_Output" , i.e. csv files are saved in the csv_images folder
def create_csv(data_d, path = "Q2_Output"):
    for i,j in data_d.items():
        f_name = path +"/"+ i.split(".")[0] + ".csv"
        f= open(f_name,"w+", newline = '')
        header = ["Entity","Count"] # Header of the csv file should be Entity and Count according to Output Format
        cw= csv.writer(f) # Creating a csv.writer object to write into the csv file
        cw.writerow(header) # writing the headers into the csv file
        cw.writerows(j.items()) # writing the Entity and Count respectively 
        cw.writerow(["Total",sum(j.values())]) # writing the total entity count in the image
        f.close()
```

```python
# Creates a dictionary having the entity names as key and their counts as values
# Takes 1 arguement ,i.e. results which is a memory-efficient generator of Results objects when using the streaming mode
def count_entities(results):
        d = {}
        names = model.names # Contains a dictionary of the labels and their entity name  
        for i,j in names.items():
            names[i] = j.capitalize() # Capitalizing the entity name for the sake of output format
            
        for r in results: # Iterating through the results generator where each iterable is a result object
            for c in r.boxes.cls: # Getting the labels/entity names in the image and iterating through it
                key = names[int(c)]
                d[key] = (d.get(key, 0)+1)
        return d
```

```python
from ultralytics import YOLO
import os
import csv

try:
    os.mkdir("Q2_Output")
except Exception as e:
    print(f"Folder already present - Error : {e}")

# Using Yolov8x.pt model since it has most parameters to modify
model = YOLO("yolov8x.pt")
directory = 'All_Images' # Directory in which all the images are in


data_d = {}
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
  # checking if it is a real file
    if os.path.isfile(f):
        # Predicting / Getting the entities in the image(filename)
        results = model.predict(source = f,conf = 0.5,save = True,save_conf = True ,stream=True)
        d = count_entities(results)
        data_d[filename] = d #Creating a dictionary where key is the image name and value is all the data of the entities within it
        
create_csv(data_d) # calling the function to create csv's for every image
```
